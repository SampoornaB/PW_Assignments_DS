### Five types of charts that we can plot using the Seaborn library :-

    1. Line Plot - A Line plot is the simplest plot. It is the visualization of a single function. This plot helps us to see the relationship between X - Axis and Y - Axis. It also takes some parameters such as hue, size, color, etc.
    
    2. Count Plot - A Count plot is used to show the counts of observations in each categorical bin using bars. This method accepts the parameters X and Y which takes the name of a variable in data, hue it is an optional parameter which helps to take column name for color encoding. Data is also an optional parameter that takes dataframe, array, or list of arrays, i.e., the dataset for plotting.
    
    3. Bar Plot - A Bar chart is a way of comparing a set of categorical data. The bar chart displays data using several bars, each representing a particular category. This method accepts the parameters X and Y  which takes the name of a variable in data, hue is an optional parameter which helps to take column name for color encoding.
    
    4. Scatter Plot - Scatter Plot is the same as a line plot, in a scatter plot instead of points being joined by line segments, the points are shown individually with a dot, circle, or any other shape. The position of each marker on the horizontal and vertical axis indicates values for an individual data point. This plot is used to observe relationships between variables. This method accepts the parameters X and Y  which takes the name of a variable in data, hue is an optional parameter which helps to take column name for color encoding. Marker is a parameter that is used to change the shape of the point which is shown in the plot.

    5. Pair Plot - Pair Plot creates a grid of axis such that each numeric variable in data will create a plot between each other the Y - Axis across a single row and the X - Axis across a single column. The diagonal plots are a univariate distribution plot that helps to draw the marginal distribution of the data in each column. A pair plots pair-wise relationships with other columns in the dataframe and also plots pairplot with itself.

```python
#Q2. Loading the "fmri" dataset and plotting a line plot

import seaborn


fmri = seaborn.load_dataset('fmri')        # Loading the 'fmri' dataset using the load_dataset() function

seaborn.lineplot(x = 'timepoint', y = 'signal', hue = 'region',  data = fmri, style = 'event')        # Plotting the line plot using the given dataset and mentioned axes
```

```python
#Q3. Loading the "titanic" dataset and plotting two box plots

import seaborn
import matplotlib.pyplot as plt


titanic = seaborn.load_dataset('titanic')        # Loading the 'titanic' dataset using the load_dataset() function

fig, axes = plt.subplots(1, 2, figsize = (16, 16))

seaborn.boxplot(x = 'pclass', y = 'age', data = titanic, ax = axes[0])        # Plotting the first box plot
seaborn.boxplot(x = 'pclass', y = 'fare', data = titanic, ax = axes[1])        # Plotting the second box plot
```

```python
#Q4. Using the "diamonds" dataset from seaborn to plot a histogram

import seaborn


diamonds = seaborn.load_dataset('diamonds')        # Loading the 'diamonds' dataset

seaborn.displot(x = 'price', hue = 'cut', data = diamonds, height = 15, aspect = 0.8)        # Plotting a price histogram with hue of 'cut' 
```

```python
#Q5. Using the "iris" dataset from seaborn to plot a pair plot

import seaborn


iris = seaborn.load_dataset('iris')        # Loading the 'iris' dataset

seaborn.pairplot(data = iris, hue = 'species')        # Plotting a pair plot with hue of 'species' 
```

```python
#Q6. Using the "flights" dataset from seaborn to plot a heatmap

import seaborn


flights = seaborn.load_dataset('flights').pivot('year', 'month', 'passengers')        # Loading the 'flights' dataset

seaborn.heatmap(flights)        # Plotting a 'flights' dataset heatmap
```
