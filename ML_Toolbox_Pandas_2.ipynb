Five of the many functions of the Pandas library are as follows :-

#### 1. value_counts() :-

This function is used to show the counts of all unique elements in columns of a dataframe. Using this function we can also view the percentages of the unique values. value_counts() doesn’t consider the NaN or the missing values by default, however we can count the NaN values by setting 'dropna = False'. This can be shown using the Titanic dataset.

```python
import pandas

titanic_df = pandas.read_csv('titanic.csv')        # Importing the titanic dataset

titanic_df['pclass'].value_counts()        # Finding the unique values under 'pclass' column and their frequencies
```

#### 2. groupby() :-

Through this function we can split and group our dataframe by certain columns to be able to view patterns and details in the data. groupby() involves 3 main steps: splitting, applying and combining. Pandas objects can be split on any of their axes. The abstract definition of grouping is to provide a mapping of labels to group names. Let us see this using the Titanic dataset.

```python
import pandas

titanic_df = pandas.read_csv('titanic.csv')        # Importing the titanic dataset

titanic_df.groupby('sex').describe()        # Using the groupby() function to segregate the unique values of sex and thus performing describe() on them
```

#### 3. set_index() :-

set_index() is a method to set a List, Series or DataFrame as index of a DataFrame. Index column can be set while making a dataframe but sometimes a DataFrame is made out of two or more DataFrames and hence later indices can be changed using this method. We can set the names of the passengers in the titanic dataset as the index as shown below.

```python
import pandas

titanic_df = pandas.read_csv('titanic.csv')        # Importing the titanic dataset

titanic_df.set_index('name')        # Making the 'name' column the index of the titanic dataset using the set_index() function
```

#### 4. loc()

loc() attribute helps us to access a group of rows and columns by label(s) or a boolean array in the given Pandas DataFrame. We can easily select only those columns where a person has survived the titanic crash using this function as shown below.

```python
import pandas

titanic_df = pandas.read_csv('titanic.csv')        # Importing the titanic dataset

titanic_df.loc[titanic_df.survived == 1]        # Finding the rows where the people survived
```

#### 5. unique() :-

The unique() function is used to get all the unique values from any feature. This is mostly used to get the categories in categorical features in the data. Unique shows all the unique values including NaNs. It treats it as a different unique value. The number of unique cabins assigned to the people can be easily found out using the unique() function on the titanic dataset as shown below.

```python
import pandas

titanic_df = pandas.read_csv('titanic.csv')        # Importing the titanic dataset

titanic_df['cabin'].unique()        # Finding the various unique values under the 'cabin' column using the unique() function
```

```python
#Q2. Function to reindex DataFrame with a new index that starts from 1 and increments by 2 for each row

import pandas

# Function to re-index a DataFrame using the given criterion

def re_indexer(Data_Frame):
    
    Data_Frame = Data_Frame.reindex(range(1, (2*len(Data_Frame) + 1), 2))
    return Data_Frame
```

```python
#Q3. Function to find the sum of the first three values in the 'Values' column in a DataFrame

import pandas

# Function to find the sum of the first 3 values under the 'Values' column

def f_3_adder(Data_Frame):
    
    return sum(Data_Frame['Values'][:3])
```

```python
#Q4. Function to find the number of words in each row of the 'Text' column in a DataFrame

import pandas

# Function to find the number of words in each row of the 'Text' column

def word_counter(Data_Frame):
    
    Data_Frame['Word_Count'] = Data_Frame['Text'].apply(lambda x : len(x.strip().split()))
    return Data_Frame
```

Pandas read_excel() function is used for reading the Excel files. Files can be imported using a URL link or can be directly imported from the disk. The Pandas Read Excel Function has various parameters.

Some of them are as follows:

* io : This parameter describes the path to the file. The string path can be any valid string. There should be a host for file URLs.

* sheet_name : In case where there are multiple sheets within the same excel file, for sheet names, strings are used. In positions on sheets with index zero, integers are used (chart sheets do not count as a sheet position). Multiple sheets can be requested using lists of strings or integers. In order to receive all worksheets, specify none.

* usecols : This parameter is used to read specific columns.

* dtype : The data type name to use for a column or data can be a dictionary using data types as values and columns as keys.

* nrows : This is used to specify the number of rows that need to be parsed.



```python
#Q7. Function to extract the usernames from the 'Email' column in a DataFrame

import pandas

# Function to extract the usernames from the 'Email' column

def u_name_extractor(Data_Frame):
    
    Data_Frame['Username'] = Data_Frame['Email'].apply(lambda x : x[:x.find('@')])
    return Data_Frame
```

```python
#Q8. Function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10

import pandas

# Function to select the required rows based on the conditions

def selector(Data_Frame):
    
    return Data_Frame[(Data_Frame['A'] > 5) & (Data_Frame['B'] < 10)]
```

```python
#Q9. Function to calculate the mean, median, and standard deviation of the values in the 'Values' column of a DataFrame

import pandas

# Function to calculate the mean, median, and standard deviation from the 'Values' column

def calc(Data_Frame):
    
    mean = Data_Frame['Values'].mean()
    median = Data_Frame['Values'].median()
    std = Data_Frame['Values'].std()
    
    return mean, median, std
```

```python
#Q10. Function to create a new column 'MovingAverage' in the DataFrame

import pandas

# Function to create a new column 'MovingAverage' in the DataFrame containing the 7 day moving average including the current day

def moving_average_7(Data_Frame):
    
    Data_Frame['Date'] = pandas.to_datetime(Data_Frame['Date'])
    Data_Frame['MovingAverage'] = Data_Frame.rolling(window = '7d', on = 'Date').Sales.mean()
    return Data_Frame
```

```python
#Q11. Function that creates a new column 'Weekday' in the DataFrame

import pandas

# Function to create a new column 'Weekday' in the DataFrame with the name of the week corresponding to the date

def weekday_finder(Data_Frame):
    
    Data_Frame['Weekday'] = Data_Frame['Date'].apply(lambda x : pandas.Timestamp(x).day_name())
    return Data_Frame
```

```python
#Q12. Function to select rows with date between '2023-01-01' and '2023-01-31' under the 'Date' column in the DataFrame

import pandas

# Function to select rows with date between '2023-01-01' and '2023-01-31' in the 'Date' column

def date_selector(Data_Frame):
    
    upper_bound = '2023-01-31'
    lower_bound = '2023-01-01'
    
    return Data_Frame[(Data_Frame['Date'] > lower_bound) & (Data_Frame['Date'] < upper_bound)]
```
