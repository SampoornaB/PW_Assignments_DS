```python
#Q1. Definition of class and object with respect to Object-Oriented Programming with example

'''
Classes are user defined templates/ blueprints for individual objects, attributes and methods.
It is a logical entitiy. To create a class we use the keyword 'class'.

'''

# Example:-

class test_1:
    
    print('A new class called test created')
    val = 19

'''
Objects are instances of a class created with specifically defined data. It is a physical entity.
We can create multiple objects of a single class. 
'''

# Example:-

obj = test_1()
print(obj.val)
```

```python
#Q3. Explaination of usage of __init__() function with example

'''
The __init__() function is a constructor which is used to initialise any parameters of the objects of a given class
'''

class Person:
    
    def __init__(self, name, age): # Initialisation
        
        self.name = name
        self.age = age
        
    def show(self):
        
        print('The name of the person is {}, and their age is {}.'.format(self.name, self.age))
    
per_1 = Person('Sampoorna', 18)

per_1.show()
```

```python
#Q4. Usage of 'self' in OOPs

'''
'Self' represents the instance of the class. By using the “self”  we can access the attributes 
and methods of the class in python. It binds the attributes with the given arguments to the current object.

Self is always pointing to the current object. Self is the first argument to be passed in constructor and instance method.
'''
```

```python
#Q5. Inheritance and its types with examples

'''
Inheritance allows us to inherit properties of a class, i.e., from a parent class to a child
class. This allows reusability of code.

Inheritance is of the following types:-

    * Single Inheritance
    * Multiple Inheritance
    * Multi-level Inheritance
    * Hierarchical Inheritance
    * Hybrid Inheritance
'''


# Example of single inheritance:-

class parent:
    
    def func_1(self):
        
        print('This is the parent class.')
        

class child(parent):
    
    def func_2(self):
        
        print('This the child class.')

obj_1 = child()

obj_1.func_1()
obj_1.func_2()

print('\n ---------------------------------------------------------- \n')


# Example of multiple inheritance:-

class parent_m:
    
    par_name_1 = ''
    
    def show_1(self):
        
        print('This is the first parent.')
        

class parent_f:
    
    par_name_2 = ''
    
    def show_2(self):
        
        print('This the second parent.')
        

class child(parent_m, parent_f):
    
    def show(self):
        
        print(f"First parent's name is {self.par_name_1} and second parent's name is {self.par_name_2}.")

obj_2 = child()

obj_2.par_name_1 = 'Ram'
obj_2.par_name_2 = 'Sita'

obj_2.show()

print('\n ---------------------------------------------------------- \n')


# Example of multi-level inheritance:-

class parent:
    
    def __init__(self, name):
        
        self.name = name
    
    def get_name(self):
        
        return self.name
    
    
class child(parent):
    
    def __init__(self, name_c, age_c):
        
        parent.__init__(self, name_c)
        self.age_c = age_c
    
    def get_age(self):
        
        return self.age_c
    
    
class grand_child(child):
    
    def __init__(self, name_gc, age_gc, address_gc):
        
        child.__init__(self, name_gc, age_gc)
        self.address_gc = address_gc
        
    def get_address(self):
        
        return self.address_gc
        

obj_3 = grand_child('Sampoorna', 18, 'Kolkata')

print('My name is {}, and my age is {}. I live in {}.'.format(obj_3.name, obj_3.age_c, obj_3.address_gc))

print('\n ---------------------------------------------------------- \n')


# Example of hierarchical inheritance:-

class parent:
    
    def show_1(self):
        
        print('This is the parent class.')
    
    
class child_1(parent):
    
    def show_2(self):
        
        print('This is child class 1.')

        
class child_2(parent):
    
    def show_3(self):
        
        print('This is child class 2.')
        
        
obj_4 = child_1()
obj_5 = child_2()

obj_4.show_1()
obj_4.show_2()
obj_5.show_1()
obj_5.show_3()

print('\n ---------------------------------------------------------- \n')


# Example of hybrid inheritance:-

class parent:
    
    def show_1(self):
        
        print("This is the parent class.")
 
 
class child_1(parent):
    
    def show_2(self):
        
        print('This is child class 1.')

        
class child_2(parent):
    
    def show_3(self):
        
        print('This is child class 2.')
        

class child_3(child_1, parent):
    
    def show_4(self):
        
        print('This is child class 3.')

              
obj_6 = child_3()

obj_6.show_1()
obj_6.show_2()
obj_6.show_4()
```
