```python
#Q1. Creating a Bokeh plot using Python code
```

Bokeh is a data visualization library in Python that provides high-performance interactive charts and plots. Bokeh output can be obtained in various mediums like notebook, html and server. It is also possible to embed bokeh plots in Django and flask apps.

Bokeh provides two visualization interfaces to users:

  * **bokeh.models :** A low level interface that provides high flexibility to application developers.
  * **bokeh.plotting :** A high level interface for creating visual glyphs.

```python
# Creating a Bokeh scatter plot using Python code

from bokeh.plotting import figure, output_notebook, show


output_notebook()

p = figure(plot_width = 500, plot_height = 500)
p.circle(x = [1, 2, 3, 4, 5], y = [4, 7, 1, 6, 3], size = 10)

show(p)
```

```python
#Q3. Customizing the appearance of a Bokeh plot, including the axes, title, and legend
```

### Customizing the appearance of a Bokeh plot

With Bokehs themes, we can quickly change the appearance of your plot. Themes are a set of pre-defined design parameters such as colors, fonts, or line styles. Bokeh comes with five built-in themes: caliber, dark_minimal, light_minimal, night_sky, and contrast. Additionally, we can define our own custom themes. To use one of the built-in themes, we assign the name of the theme we want to use, to the theme property of our document.

### Customizing the plot axes

We can set various attributes to change the way the axes in your plot work and look.

Options for customizing the appearance of your plot include:

  * Setting labels for our axes
  * Styling the numbers displayed with your axes
  * Defining colors and other layout properties for the axes themselves

### Adding a title to the plot and customizing it

We can add a title to the plots by passing the title argument to the figure() function.

### Adding a legend to the plot and customizing it

Bokeh automatically adds a legend to our plot if we include the legend_label attribute when calling the renderer function.

```python
from bokeh.io import curdoc
from bokeh.plotting import figure, show
import random


curdoc().theme = 'light_minimal'    # Applying a theme to current document

plot = figure(max_width = 500, height = 500, sizing_mode = 'stretch_width', title = 'X vs Y Plot')    # Applying a title to the plot

x = list(range(0, 25))
y = random.sample(range(0, 100), 25)

plot.line(x, y, line_color = 'black')
plot.circle(x, y, size = 10, color = 'white', line_color = 'black', legend_label = 'Data points')    # Applying a legend label to the plot

# Customizing the title

plot.title.align = 'center'
plot.title.text_font_size = '25px'
plot.title.text_color = 'red'

# Customizing the x - axis

plot.xaxis.axis_label = 'X - Axis'
plot.xaxis.axis_line_width = 3
plot.xaxis.axis_line_color = 'red'

# Customizing the y - axis

plot.yaxis.axis_label = 'Y - Axis'
plot.yaxis.major_label_text_color = 'orange'
plot.yaxis.major_label_orientation = 'horizontal'

# Customizing the appearance of the legend

plot.legend.title = 'Legend Title'
plot.legend.label_text_font = 'times'
plot.legend.label_text_font_style = 'italic'
plot.legend.label_text_color = 'purple'

show(plot)
```

```python
#Q4. Defining Bokeh server and using it to create interactive plots that are updated in real time
```

### Bokeh Server

The Bokeh server is a component of Bokeh that allows you to build interactive web applications that are connected to Python code running on a server. Bokeh server makes it easy to create interactive web applications that connect front-end UI events to running Python code. Bokeh creates high-level Python models, such as plots, ranges, axes, and glyphs, and then converts these objects to JSON to pass them to its client library, BokehJS.

### Using Bokeh Server to create interactive plots that are updated in real time

The most flexible way to create interactive data visualizations with the Bokeh server is to create Bokeh applications and serve them with the bokeh serve command. The Bokeh server then uses the application code to create sessions and documents for all connecting browsers. The Bokeh server executes the application code with every new connection and creates a new Bokeh document, syncing it to the browser. The application code also sets up the callbacks that should run whenever properties, such as widget values, change.

```python
#Q5. Embedding a Bokeh plot into a web page or dashboard using Flask or Django
```

We can embed Bokeh apps so that every page load either creates and displays a new session and document or outputs a specific, existing session. If an application is running on a Bokeh server that makes it available at some URL, we will typically want to embed the entire application in a web page. This way, the page will create a new session and display it to the user every time it loads. We can achieve this with the server_document() function. This function accepts the URL to a Bokeh server application and returns a script that embeds a new session from that server every time the script executes.

We cab create a basic bokeh graph and embed it into our Flask or Django web app. For that, we import components from bokeh.embed which returns the individual components. The function bokeh.embed.components() returns a script that contains that data for your plot with a \<div\> tag in which the plot view is loaded.
