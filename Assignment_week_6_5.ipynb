### GET Method :-

It is the most common method which can be used to send data in an unencrypted form to the server. Whenever this method is used a GET message is send, and the server returns data. For example, in an eCommerce application, there is a search button to search for an item. After entering a keyword when we notice the URL, the keyword we searched for gets displayed in the URL. This request is useful when the data is not sensitive.

### POST Method :-

It is used to send HTML form data to the server. The data received by the POST method is not cached by the server. While logging into our gmail account, the registration and login form is the best example of the post method. Whenever we enter the required information and submit the data, it gets transferred over to the POST request. If we notice the URL, unlike GET request, no information will be displayed in the URL.

The Request, in Flask, is an object that contains all the data sent from the Client to Server. This data can be recovered using the GET/POST Methods. In order to process the request data, it should be imported from the Flask module.

Some of the important attributes of request object are as followes :−

* Form − It is a dictionary object containing key and value pairs of form parameters and their values.

* args − parsed contents of query string which is part of URL after question mark (?).

* Cookies − dictionary object holding Cookie names and values.

* files − data pertaining to uploaded file.

* method − current request method.

```python
#Q3. Use of redirect() in Flask
```

Flask class provides the redirect() function which redirects the user to some specified URL with the specified status code. An HTTP status code is a response from the server to the request of the browser. When we visit a website, a request is sent to the server, and the server then responds to the browser's request with a three-digit code: the HTTP status code. 302 is the default status code.

#### Syntax for Flask redirect() :-

    Flask.redirect(location, statuscode, response)

In the above function −

* <b>location</b> parameter is the URL where response should be redirected.
* <b>statuscode</b> sent to browser’s header, defaults to 302.
* <b>response</b> parameter is used to instantiate response.

```python
#Q4. Defining templates and the use of render_template() in Flask
```

Templates are files that contain static data as well as placeholders for dynamic data. A template is rendered with specific data to produce a final document. In Flask, we can use the Jinja templating language to render HTML templates. Jinja allows us to respond with an HTML template where we can use many features that are not available in standard HTML, such as variables, if statements, for loops, filters, and template inheritance. These features allow us to efficiently write easy-to-maintain HTML pages.

#### Use of render_template() :-

The render_template() method is used to generate output from a template file based on the Jinja2 engine that is found in the application's templates folder. It is a key feature of Flask's template engine, which allows us to separate the presentation logic (HTML templates) from the application logic (Python code). The render_template() function takes the name of the template file as its first argument and optional keyword arguments that represent variables to be passed to the template.

```python
from flask import Flask, request, jsonify

app = Flask(__name__)


@app.route("/", methods=['POST'])
def home_screen():
    return '<h1 style="text-align:center;">HELLO \U0001f600</h1>'


@app.route('/postman_test', methods=['POST'])
def logic_ops():
    if request.method == 'POST':

        item_type = request.json['item_type']
        price = int(request.json['price'])

        if item_type == 'food_beverages':
            discount = 0.2 * price
            return jsonify('Your payable amount is ' + str(price - discount) + ' with a discount amount of ' + str(discount) + '.')
        
        elif item_type == 'clothing':
            discount = 0.1 * price
            return jsonify('Your payable amount is ' + str(price - discount) + ' with a discount amount of ' + str(discount) + '.')
        
        elif item_type == 'transportation':
            discount = 0.3 * price
            return jsonify('Your payable amount is ' + str(price - discount) + ' with a discount amount of ' + str(discount) + '.')
        
        elif item_type == 'housing':
            discount = 0.4 * price
            return jsonify('Your payable amount is ' + str(price - discount) + ' with a discount amount of ' + str(discount) + '.')
        
        elif item_type == 'others':
            discount = 0.05 * price
            return jsonify('Your payable amount is ' + str(price - discount) + ' with a discount amount of ' + str(discount) + '.')


if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
